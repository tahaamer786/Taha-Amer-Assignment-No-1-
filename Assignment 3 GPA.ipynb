### Step 4: Handling the Missing Values

### To ensure that data is clean and complete, we handled the missing values using following approach.

#### . Numerical columns like Attendence Rate, sleep hours per night and online course work. Engagement were filled with their mean values.
#### . The categorial columns Major was filled with its most frequent values (mode).

This ensures that our data set has no any missing values. Making it ready for feature encoding and model training.


```python
## Fill missing numerical columns with mean
df['Attendance Rate'].fillna(df['Attendance Rate'].mean(), inplace=True)
df['Online Coursework Engagement'].fillna(df['Online Coursework Engagement'].mean(), inplace=True)
df['Sleep Hours per Night'].fillna(df['Sleep Hours per Night'].mean(), inplace=True)

## Fill missing categorial columns 'Major with the most frequent values' (mode).
df['Major'].fillna(df['Major'].mode()[0], inplace=True)

## confrim all missing values are handled
print("Missing values after cleaning")
print(df.isnull().sum())
```

```python
## Drop the rows with missing values
df_clean = df.dropna()

## Separate features and Target
X = df_clean.drop(columns=['College GPA'])
Y = df_clean['College GPA']                           
```

```python
## Encode categorail variables
X = pd.get_dummies(X, drop_first=True)
```

```python
print("Features shape:", X.shape)
print("Target shape:", Y.shape)
print("Preview of Encoded Features:")
print(X.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)

print("Training set size:" , X_train.shape)
print("Testing set size:" , X_test.shape)
```

## Linear Regression

#### . Since our target variable college GPA is continues , we use linear regression instead of logic regression
#### . This model predicts GPA
#### . Evaluation through metrices


## Some Errors
## MAE: Average Absolute Error
## MSE: Average Squared Error
## RMSE: Root of Squared Error

```python
from sklearn.linear_model import LinearRegression

## Initiliaze and train the model
lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

## Support Vector Regression

#### . SVR is a Regression version of support 
#### . It fits the best function within a certain margin to predict continous values
#### . Good for capturing non linear patterns 

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

## Make Predictions
Y_pred_lin = lin_model.predict(X_test)

## Evaluate 
mae_lin = mean_absolute_error(Y_test, Y_pred_lin)
mse_lin = mean_squared_error(Y_test, Y_pred_lin)
rmse_lin = np.sqrt(mse_lin)

print("Linear Regression Evaluation")
print("MAE:" , mae_lin)
print("MSE:" , mse_lin)
print("RMSE:", rmse_lin)
```

## Stochastic Gradient Descent (SGD) Regressor
#### . Predict output (e.g., GPA)
#### .Measure the error (how far off the guess is)
#### .Update the model to reduce that error using a small step in the direction that lowers the error

```python
from sklearn.linear_model import SGDRegressor

## Initialize the SGD Regressor
sgd_model = SGDRegressor(max_iter=1000, tol=1e-3, random_state=42)

## Train the model 
sgd_model.fit(X_train, Y_train)

## predict
y_pred_sgd = sgd_model.predict(X_test)

## Evaluate
mae_sgd = mean_absolute_error(Y_test, y_pred_sgd)
mse_sgd = mean_squared_error(Y_test, y_pred_sgd)
rmse_sgd = np.sqrt(mae_sgd)

print ("SGD Regressor Evaluation:")
print("MAE:", mae_sgd)
print("MSE:", mse_sgd)
print("RMSE:", rmse_sgd)
```

## Multi-Layer Perceptron (MLP) Regressor

#### . Feed features (e.g., hours studied, grades) into the input layer.
#### . Pass values through hidden layers using activation functions (like ReLU or tanh).
#### . Output a predicted GPA.
#### . Do this for many epochs until the model learns a good mapping.



```python
from sklearn.neural_network import MLPRegressor

## Initialize the MLP Regressor
mlp_model = MLPRegressor(hidden_layer_sizes=(100, 100), max_iter=1000, random_state=42)

## Train the model
mlp_model.fit(X_train, Y_train)

## Predict 
y_pred_mlp = mlp_model.predict(X_test)

## Evaluate
mae_mlp = mean_absolute_error(Y_test, y_pred_mlp)
mse_mlp = mean_squared_error(Y_test, y_pred_mlp)
rmse_mlp = np.sqrt(mse_mlp)


print ("MLP Regressor Evaluation:")
print("MAE:", mae_mlp)
print("MSE:", mse_mlp)
print("RMSE:", rmse_mlp)
```

## Decision Tree Regression

#### .The model asks yes/no questions (e.g., Is attendance > 80%?).
#### .Based on the answer, it splits the dataset into smaller groups.
#### . Each split tries to reduce prediction error.
#### .At the end (the leaf nodes), it assigns a value (e.g., average GPA of that group).

```python
from sklearn.tree import DecisionTreeRegressor

## Initialize the Decision Tree Regressor
dt_model = DecisionTreeRegressor(random_state=42)

## Train the model 
dt_model.fit(X_train, Y_train)

## Predict
Y_pred_dt = dt_model.predict(X_test)

## Evaluate 
mae_dt = mean_absolute_error(Y_test, Y_pred_dt)
mse_dt = mean_squared_error(Y_test, Y_pred_dt)
rmse_dt = np.sqrt(mse_dt)

print("Decision Tree Regressor Evaluation:")
print("MAE:", mae_dt)
print("MSE:", mse_dt)
print("RMSE:", rmse_dt)
```
